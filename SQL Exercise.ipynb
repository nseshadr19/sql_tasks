```python
import sqlite3
from sqlite3 import Error

 
def create_connection(db_file):
    """ create a database connection to the SQLite database
        specified by the db_file
    :param db_file: database file
    :return: Connection object or None
    """
    conn = None
    try:
        conn = sqlite3.connect(db_file)
        print(sqlite3.version)
    except Error as e:
        print(e)
 
    return conn

 
def select_all_tasks(conn):
    """
    Query all rows in the tasks table
    :param conn: the Connection object
    :return:
    """
    cur = conn.cursor()
 
    
    query1 = """
        SELECT Facility_Name, Revenue
        FROM (

        SELECT f.name AS Facility_Name, SUM(
        CASE WHEN b.memid =0
        THEN (
        slots * guestcost
        )
        ELSE (
        slots * membercost
        )
        END ) AS Revenue
        FROM Bookings b
        JOIN Facilities f ON b.facid = f.facid
        JOIN Members m ON m.memid = b.memid
        GROUP BY facility_name
        )x
        WHERE Revenue <1000
        ORDER BY 2 DESC
        """
    cur.execute(query1)
 
    rows = cur.fetchall()
 
    for row in rows:
        print(row)      
        
    query2 = """
        SELECT 
        r.surname|| ','|| r.firstname AS Member,
         m.surname|| ','|| m.firstname AS RecommendedBy
        FROM
            Members r
        INNER JOIN Members m ON 
            r.recommendedby=m.memid 
        where m.memid<>0
        order by 1
        """
    cur.execute(query2)
 
    rows = cur.fetchall()
 
    for row in rows:
        print(row)  
        
    query3 = """
    SELECT f.name AS Facility,  m.firstname|| ' '|| m.surname  AS Member, COUNT( starttime ) AS Count_usage
    FROM `Members` m
    JOIN Bookings b ON m.memid = b.memid
    JOIN Facilities f ON b.facid = f.facid
    WHERE m.memid <>0
    GROUP BY f.name,  m.firstname|| ' '|| m.surname 
    ORDER BY 1
        """
    cur.execute(query3)
 
    rows = cur.fetchall()
 
    for row in rows:
        print(row)
        
    query4 = """
        SELECT MONTH(starttime) || '-'|| YEAR(starttime) AS
        MONTH , f.name AS Facility, COUNT( starttime ) AS Count_usage
        FROM Members m
        JOIN Bookings b ON m.memid = b.memid
        JOIN Facilities f ON b.facid = f.facid
        WHERE m.memid <>0
        GROUP BY MONTH(starttime) || '-'|| YEAR(starttime) , f.name

        """
    cur.execute(query4)
 
    rows = cur.fetchall()
 
    for row in rows:
        print(row)
    
        
        
    


def main():
    database = "sqlite_db_pythonsqlite.db"
 
    # create a database connection
    conn = create_connection(database)
    with conn: 
        print("2. Query all tasks")
        select_all_tasks(conn)
 
 
if __name__ == '__main__':
    main()
```
